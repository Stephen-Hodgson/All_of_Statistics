The pdf of y is:

$f_Y(y)= \begin{cases}
\frac{1}{y\sqrt{y}\sqrt{2\pi}} \exp{(-\frac{(\ln{y})^2}{2})} &y>0\\\\
0 &\text{elsewhere}
\end{cases}$


```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.linspace(0.01,10,10000)
y=np.exp(-np.log(x)**2/2)/(x*np.sqrt(2*np.pi))
```

```python
plt.plot(x,y)
plt.show()
```

```python
ns=np.random.randn(10000)
ys=np.exp(ns)
```
