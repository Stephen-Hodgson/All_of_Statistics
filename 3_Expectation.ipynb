```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import cauchy
```

```python
n=10000

i=np.arange(1,n+1)
norm_xi=np.random.randn(n)
norm_mean=np.cumsum(norm_xi)/i

plt.plot(i,norm_mean)
plt.show()
```

```python
cauchy_xi = cauchy.rvs(size=n)
cauchy_mean = np.cumsum(cauchy_xi)/i

plt.plot(i,cauchy_mean)
plt.show()

```

```python
n = 10000
m=10

i=np.arange(1,n+1)
yi = np.random.choice([-1, 1], p=[0.5, 0.5], size=(m,n))
xis=np.cumsum(yi, axis=1)

for xi in xis:
    plt.plot(i, xi)
plt.show()
```

```python
import seaborn as sns
```

```python
m=10000
ns = [1,5,25,100]
means=[]
vars=[]


for n in range(1, 101):
    xis=np.random.uniform(0,1,size=(m,n))
    mean=np.sum(xis, axis = 1)/n
    means.append(np.mean(mean))
    vars.append(np.var(mean))

    if n in ns:
        print("n = {}, mean = {}, variance in mean = {}".format(n, np.mean(mean), np.var(mean)))
        sns.histplot(mean, stat='density')

plt.xlabel('Xbar')
plt.show()
```

```python
i = np.arange(1,101)

plt.plot(i,means)
plt.xlabel('n')
plt.ylabel('average of Xbar')
plt.show()

plt.plot(i, 1 / (12 * i), label = 'calculated')
plt.plot(i, vars, label = 'measured')
plt.yscale('log')
plt.xlabel('n')
plt.ylabel('variance of xbar')
plt.legend()

plt.show()
```
