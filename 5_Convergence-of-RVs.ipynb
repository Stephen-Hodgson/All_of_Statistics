```python
import numpy as np
from scipy.stats import poisson
from scipy.stats import norm
```

```python
n=100
N=10000

xis = poisson.rvs(1, size=(N,n))
xbars = np.mean(xis, axis=1)

outside = (xbars<0.9)
prop = np.mean(outside)
print(prop)
```

```python
n=100
N=10000

xis = norm.rvs(1, 1, size=(N,n))
xbars = np.mean(xis, axis=1)

outside = (xbars<0.9)
prop = np.mean(outside)
print(prop)
```
