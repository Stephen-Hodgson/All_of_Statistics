Here I compare the estimators from the following two exercises:

```python
theta = np.random.uniform(1, 100)
print("theta = ", theta)
```

Although theta1 is biased, it is consistent and has a much lower standard error than theta2

```python
n = 10000
xis = np.random.uniform(0, theta, size=n)
xbar = xis.mean()

theta1 = max(xis)
theta2 = 2*xbar

print("theta = {}\nmax(xis) = {}\n2xmean = {}".format(theta, theta1, theta2))
```
