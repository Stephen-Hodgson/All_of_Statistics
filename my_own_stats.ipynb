```python
import numpy as np
from scipy.stats import uniform
```

```python
N=10000000
X = uniform.rvs(size=N)
Y = np.random.uniform(low=X)
```

```python
print(np.mean(Y))
```

```python
X1 = uniform.rvs(size=N)
X2 = uniform.rvs(size=N)
M = np.maximum(X1, X2)
print(np.mean(M))
```

```python
X1 = uniform.rvs(size=N)
X2 = uniform.rvs(size=N)
M = np.maximum(X1, X2)
m = np.minimum(X1, X2)
result = np.maximum(m, M-m, 1-M)
print(np.mean(result))
```

```python
results = []

N=1000
n=1000

for _ in range(N):
    balls = [0, 1]
    for i in range (n):
        balls.append(np.random.choice(balls))
    results.append(np.sum(balls))
```

```python
plt.hist(results, bins=50)
plt.show()
```
