The pdf of y is:

$f_Y(y)= \begin{cases}
\frac{1}{y\sqrt{y}\sqrt{2\pi}} \exp{(-\frac{(\ln{y})^2}{2})} &y>0\\\\
0 &\text{elsewhere}
\end{cases}$


```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0.01, 10,10000)
y = np.exp( -np.log(x) ** 2 / 2) / (x * np.sqrt(2 * np.pi))
```

```python
plt.plot(x,y)
plt.show()
```

```python
ns = np.random.randn(10000)
ys = np.exp(ns)
```

```python
sns.histplot(ys, stat='density')
plt.plot(x, y)
plt.xlim([0, 10])
plt.show()
```

```python
beta = 1

ui = np.random.uniform(0,1,size=10000)
xi = -beta * np.log(1 - ui)
sns.histplot(xi, stat='density')

x = np.linspace(0, 10)
y = np.exp(-x / beta) / beta

plt.plot(x, y)
plt.show()
```

```python
a = norm.cdf(7, loc=3, scale=4)
print('a = ', '%.3f' % a)

b = 1 - norm.cdf(-2, loc=3, scale=4)
print('b = ', '%.3f' % b)

c = norm.ppf(0.95, loc=3, scale=4)
print('c = ', '%.3f' % c)

d = norm.cdf(4, loc=3, scale=4) - norm.cdf(0, loc=3, scale=4)
print('d = ', '%.3f' % d)

e = norm.ppf(0.975, loc=3, scale=4)
print('e = ', '%.3f' % e)
```
