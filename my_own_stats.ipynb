```python
import numpy as np
from scipy.stats import uniform
```

```python
N=10000000
X = uniform.rvs(size=N)
Y = np.random.uniform(low=X)
```

```python
print(np.mean(Y))
```

```python
X1 = uniform.rvs(size=N)
X2 = uniform.rvs(size=N)
M = np.maximum(X1, X2)
print(np.mean(M))
```

```python
X1 = uniform.rvs(size=N)
X2 = uniform.rvs(size=N)
M = np.maximum(X1, X2)
m = np.minimum(X1, X2)
result = np.maximum(m, M-m, 1-M)
print(np.mean(result))
```
