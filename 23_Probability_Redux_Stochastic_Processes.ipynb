```python
import numpy as np
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
```

```python
a, b = 0.1, 0.3
P = np.array([[1-a, a], [b, 1-b]])
pi = np.array([b / (a + b), a / (a + b)])
Pinf = np.array([pi, pi])
```

```python
mu0 = np.array([0.5, 0.5])
N = 10000
```

```python
mus = (mu0 @ [np.linalg.matrix_power(P, x) for x in range(N)])
```

```python
from scipy.stats import bernoulli
```

```python
Xi = bernoulli.rvs(mus[:, 1])
i = np.arange(1, N+1)
```

```python
cummean = np.cumsum(Xi)/i
```

```python
plt.plot(i, 1 - cummean, label = 'p1')
plt.plot(i, cummean, label = 'p2')
plt.hlines([0.25, 0.75], xmin = 1, xmax = N, linestyles = '--', color='k')
plt.legend()
plt.xscale('log')
plt.show()
```
